```python
#if statements
age=18
if age>=18:
    print("eligible to vote")
```

```python
#if else statements
num=9
if num%2==0:
    print('even')
else:
    print('odd')
```

```python
#if elif else statements
marks=85
if marks>=90:
    print('grade A')
elif marks>=75:
    print('grade B')
else:
    print('grade C')
```

```python
#for loop
for i in range(1,6):
    print(i)
```

```python
#while loop
i=1
while i<=5:
    print(i)
    i+=1
```

```python
#break statement
for i in range(1,6):
    if i==3:
        break
    print(i)
```

```python
#continue statement
for i in range(1,6):
    if i==3:
        continue
    print(i)
```
